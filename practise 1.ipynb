```python
num_pests = int(input("Enter number of pests per plant: "))
damage_percentage = int(input("Enter crop damage percentage: "))
temperature = int(input("Enter temperature (°C): "))
humidity = int(input("Enter humidity level (%): "))

severity_level = ""
recommendation = ""
cost_of_intervention = ""

# Classify Severity
if num_pests < 5 and damage_percentage < 10:
    severity_level = "LOW"
    recommendation = "Routine monitoring. Consider preventative measures like natural predators."
    cost_of_intervention = "₹0/hectare"
elif (5 <= num_pests <= 15) or (10 <= damage_percentage <= 25):
    severity_level = "MEDIUM"
    recommendation = "Monitor daily. Consider targeted pesticide spray if severity increases."
    cost_of_intervention = "₹500/hectare"
elif num_pests > 15 or damage_percentage > 25:
    severity_level = "HIGH"
    recommendation = "Immediate intervention required. Apply broad-spectrum pesticides and consult an expert."
    cost_of_intervention = "₹1500/hectare"
else:
    severity_level = "UNKNOWN"
    recommendation = "Cannot determine severity based on inputs."
    cost_of_intervention = "N/A"

print(f"Pests per plant: {num_pests}, Damage: {damage_percentage}%, Temp: {temperature}°C, Humidity: {humidity}%")
print(f"Severity Level: {severity_level}")
print(f"Recommendation: {recommendation}")
print(f"Cost of intervention: {cost_of_intervention}")
```

```python
crop_type = input("Enter crop type (Rice, Wheat, Cotton): ").capitalize()
days_since_sowing = int(input("Enter days since sowing: "))

maturity_periods = {
    "Rice": (120, 150),
    "Wheat": (140, 180),
    "Cotton": (160, 200)
}

status = ""
m_range = ""
recommendation = ""
yield_loss_info = ""

if crop_type in maturity_periods:
    min_days, max_days = maturity_periods[crop_type]
    m_range = f"{min_days}-{max_days} days"

    if min_days <= days_since_sowing <= max_days:
        status = "READY FOR HARVEST"
        recommendation = "Begin harvesting within next 2 weeks for optimal yield."
        yield_loss_info = "Expected yield loss if delayed: 2% per day"
    elif days_since_sowing < min_days:
        status = "NOT READY YET"
        days_to_maturity = min_days - days_since_sowing
        recommendation = f"Continue monitoring. Crop will be ready in approximately {days_to_maturity} days."
        yield_loss_info = ""
    else: # days_since_sowing > max_days
        status = "OVERDUE FOR HARVEST"
        overdue_days = days_since_sowing - max_days
        potential_loss = overdue_days * 2 # Assuming 2% loss per day overdue
        recommendation = "Harvest immediately to minimize further losses."
        yield_loss_info = f"Potential yield loss: {potential_loss}% already incurred due to delay."
else:
    status = "UNKNOWN CROP TYPE"
    recommendation = "Please enter a valid crop type (Rice, Wheat, Cotton)."
    yield_loss_info = ""

print(f"Crop: {crop_type}, Days since sowing: {days_since_sowing}")
print(f"Status: {status}")
if m_range:
    print(f"Maturity range: {m_range}")
print(f"Recommendation: {recommendation}")
if yield_loss_info:
    print(f"Expected yield loss if delayed: {yield_loss_info}")
```
