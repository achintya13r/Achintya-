```python
rainfall = float(input("Enter rainfall (mm): "))
temperature = float(input("Enter temperature (°C): "))
soil = float(input("Enter soil quality (80-100): "))

base_yield = 40
total_change = 0
for factor in range(3):
    for step in range(1):
        # Rainfall factor
        if factor == 0:
            if rainfall > 600:
                total_change += 2 * ((rainfall - 600) // 10)
            elif rainfall < 400:
                total_change -= 2 * ((400 - rainfall) // 10)

        # Temperature factor
        elif factor == 1:
            if temperature < 25:
                total_change -= (25 - temperature)
            elif temperature > 30:
                total_change -= (temperature - 30)

        # Soil quality factor
        elif factor == 2:
            if soil < 80:
                total_change -= (80 - soil)
            elif soil > 100:
                total_change -= (soil - 100)

final_yield = base_yield * (1 + total_change / 100)
print("Total percentage change:", total_change, "%")
print("Predicted crop yield:", final_yield, "quintals per hectare")
```
